## Data Analysis: USDCAD Timeseries Analysis 

This Project focuses on the Time Series analysis of USDCAD. Futhermore, how the price and return behavior of US 10 year Treasury bonds, Canadian 10 year treasury bonds and WTI Crude oil impact the returns of USDCAD. In this data analysis project, we download monthly data from FRED API and conduct multiple linear regression and regime switching analysis to determine if there any specific relationship between the variables and if so, how do they change over time? The motivation for this study is that the canadian dollar is known to be a commodity currency. The price of WTI crude oil and other commodities can greatly impact the canadian economy. In this analysis, we will explore if the monthly returns of crude oil have a significant impact on the price behavior of USDCAD. We will also add interest rates of 10 year bonds to improve the regression and just for fun. 

## Importing and cleaning of Timeseries Data

```python
# Importing necessary packages for Canadian Dollar/Us Dollar Analysis 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import pandas_datareader # To import FRED data
import datetime
from arch.unitroot import DFGLS, ADF, KPSS, PhillipsPerron # Downloaded package using 'pip install arch' for Stationarity tests
import statsmodels.formula.api as smf # For Multivariable OLS Regression
from statsmodels.regression.rolling import RollingOLS # For Rolling Regression
import statsmodels.api as sm
```

```python
# a. Importing necessary data from FRED from January 1990 to April 2020
start = datetime.datetime(1990,1,1)
end = datetime.datetime(2020,4,1)
USDCAD = pandas_datareader.data.DataReader('DEXCAUS','fred', start, end) # USDCAD Exchange Rate
Oil = pandas_datareader.data.DataReader('DCOILWTICO','fred', start, end) # North American Crude Oil
UST = pandas_datareader.data.DataReader('DGS10','fred', start, end) # 10-year US Treasury yield 
CAT = pandas_datareader.data.DataReader('IRLTLT01CAM156N','fred', start, end) # 10-year Canadian Treasury yield
```

```python
# b. Merging Dataframes
complete = pd.merge(CAT, Oil, left_index=True, right_index=True, how='left')
complete = pd.merge(complete, UST, left_index=True, right_index=True, how='left')
complete = pd.merge(complete, USDCAD, left_index=True, right_index=True, how='left')
```

```python
# c. Renaming Variables and drop missing variables
Columnrename={'IRLTLT01CAM156N':'CAD_10_Year_Treasury', 'DCOILWTICO':'WTI_CRUDE','DGS10':'US_10_Year_Treasury','DEXCAUS':'USDCAD'}
complete.rename(columns = Columnrename, inplace=True)
complete = complete.dropna()
```

```python
# d. Create new return variables 
Column_Variables=['CAD_10_Year_Treasury', 'WTI_CRUDE','US_10_Year_Treasury', 'USDCAD']

for columns in Column_Variables:
    complete[columns +'_pct'] = complete[columns].pct_change() # Percent change of each of the original variables 
complete = complete.dropna()
complete.head()
```

## Data Analysis


Now we have observed some interesting behavior in our variables. We will conduct multi-variable OLS regression and rolling window regression to see if our variables adjust over time, highlighting possible regime shifts that influnce USDCAD since the 1990s.

Before conducting our regressions, it would be useful to construct a correlation matrix in order to observe if there are any interesting correlations.

```python
# a. Correlation Matrix of our variables
complete.corr()
```

Next, we will conduct a regression on how US 10 year treasury yield, Canadian 10 year treasury yield and Crude oil returns impact the returns on CAD/USD. However, it is essential to check that our manipulated variables are now stationariy prior to conducting our regression. we confirm that the new percent change variables are stationary with a Augmented Dickey-Fuller Test

```python
# b. Conduct Stationary test via Augmented Dickey fuller test 

# Create function to test for stationarity 
def stationarity_test(df):
    '''
    Determines if series from a dataframe is stationary by conducting a unit root test, the augmented dickey fuller test.
    
    Arguments:
    df(series): series to be tested for stationarity 
    
    Returns: if series is stationary or not and if stationary with what level of significance.
    '''
    
    result = ADF(df)
    
    if abs(result.stat) > abs(result.critical_values['5%']) and abs(result.stat) < abs(result.critical_values['1%']):
        x ='Stationary with 5% level of significance'
    elif abs(result.stat) > abs(result.critical_values['1%']):
        x='Stationary with 1% level of significance'
    else:
        x='Non-Stationary'
    
    return x

#  Test for Stationarity for all of our variables
updated_list=['CAD_10_Year_Treasury_pct', 'WTI_CRUDE_pct','US_10_Year_Treasury_pct','USDCAD_pct']
old_list=['CAD_10_Year_Treasury', 'WTI_CRUDE','US_10_Year_Treasury','USDCAD']

for columns in old_list:
    print(columns+':'+stationarity_test(complete[columns]))

for columns in updated_list:
    print(columns+':'+stationarity_test(complete[columns]))
```

We can conclude that $\underline{ALL}$ of our time series variables converted to percent returns surpass the 1% critical value of possessing a unit root. Therefore, we can assume stationarity and proceed with a multivariable regression. 

```python
# c. Conduct Multivariable OLS Regression
formula = 'USDCAD_pct ~ CAD_10_Year_Treasury_pct + US_10_Year_Treasury_pct + WTI_CRUDE_pct'
results = smf.ols(formula, complete).fit()
print(results.summary())
```

From our first regression, we discover that Canadian 10 year treasury yield percent returns and WTI crude oil percent returns are highly significant both surpassing the 5% critical level. However, US 10 year treasury yields appear to have no significant effect on USDCAD returns. Furthermore, the sign of our signifiant coeffecients both are negative. This makes logical sense from theory and expectations. As Canadian interest rates rise holding all other variables constant, there will be more capital flows into Canada causing an appeciation of the Canadian dollar. As crude oil prices increases, the large oil sector of the canadian economy will profit from increased prices causing larger GDP growth. As output increases in Canada, this will result in higher money demand causing the canadian dollar to appreciate. It is surprising that the US 10 year treasury yield has no significant impact on USDCAD returns. This requires more study. 

```python
formula = 'USDCAD_pct ~ CAD_10_Year_Treasury_pct+ WTI_CRUDE_pct'
results = smf.ols(formula, complete).fit()
print(results.summary())
```

We redid the OLS regression removing our insiginificant variable. This pushed all of our variables to surpass the 1% level of significance and improved our adjusted r-squared. One important thing to note from our regression is that even though are remaining variables are highly significant, they only account for 17.3 percent of the variation in USDCAD returns. This OLS regression has lots of room for improvements. 

```python
# d. Conducting a rolling regression 

endog = complete['USDCAD_pct'] # Endogenous Variable-USDCAD Return
exog_vars = ['CAD_10_Year_Treasury_pct', 'WTI_CRUDE_pct'] # Our remaining significant exogenous variables
exog = sm.add_constant(complete[exog_vars])
rols = RollingOLS(endog, exog, window=60) # Conduct Rolling windowed OLS regression with a window lenght of 60 months 
rres = rols.fit()
fig = rres.plot_recursive_coefficient(variables=exog_vars, figsize=(14,18))
```
